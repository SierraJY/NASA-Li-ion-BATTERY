```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
real = pd.read_csv("../DATA/DataFrame/B0005_capacity_df.csv")
FNN10 = pd.read_csv("./FNN_10N.csv")
FNN40 = pd.read_csv("./FNN_40N.csv")
CNN10K5K = pd.read_csv("./CNN_10K_5K.csv")
CNN30K15K = pd.read_csv("./CNN_30K_15K.csv")
LSTM = pd.read_csv("./LSTM.csv")
```

```python
plt.plot(range(len(real.index)), real["capacity"], label="REAL", color="Black")
plt.plot(range(len(real.index)), FNN10.iloc[:,0], label="FNN-10neu")
plt.plot(range(len(real.index)), FNN40.iloc[:,0], label="FNN-40neu")
plt.plot(range(len(real.index)), CNN10K5K.iloc[:,0], label="CNN-10k-5k")
plt.plot(range(len(real.index)), CNN30K15K.iloc[:,0], label="CNN-30k-15k")
plt.plot(range(len(real.index)), LSTM.iloc[:,0], label="LSTM")

plt.ylim([1.2, 2.2])
plt.title("Use only Voltage")
plt.xlabel(xlabel='Cycle', labelpad=15)
plt.ylabel(ylabel='Capacity', labelpad=15)
plt.legend()
plt.show()
```
